# VISUALLIZING WORLD DATA

### Confirmed cases (Total Cases) all over the world

```python
data_World_over_time = data_World.groupby(["date"])[["total_cases","new_cases","total_deaths","new_deaths"]].sum().reset_index().sort_values("date",ascending=True).reset_index(drop=True)
data_World_over_time
```

### Using line-plot to calcolat the Total deaths cases evolution over time¶ 

```python
plt.figure(figsize=(40,20))
plt.plot(data_World_over_time.index, data_World_over_time['total_cases'])
plt.title('Evolution of Confirmed Covid-19 cases over time in the word', fontsize=50)
plt.xlabel('Days', fontsize=80)
plt.ylabel('Confirmed cases', fontsize=80)
```

### New cases all over the world

```python
plt.figure(figsize=(20,10))
plt.bar(data_World_over_time.index, data_World_over_time['new_cases'])
plt.title('Evolution of Covid-19 New Cases over time in the world', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
```

### Putting it all together

```python
plt.figure(figsize=(20,10))
plt.plot(data_World_over_time.index, data_World_over_time['total_cases'], label='Confirmed')
plt.plot(data_World_over_time.index, data_World_over_time['total_deaths'], label='Total Deaths')
plt.plot(data_World_over_time.index, data_World_over_time['new_cases'], label='New')
plt.plot(data_World_over_time.index, data_World_over_time['new_deaths'], label='New Deaths')
plt.legend(loc=0)
plt.show()
```
